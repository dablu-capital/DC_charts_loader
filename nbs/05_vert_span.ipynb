```python
from pathlib import Path
import pandas as pd
import sys

ROOT_DIR = Path().cwd().parent

if str(ROOT_DIR) not in sys.path:
    sys.path.insert(0, str(ROOT_DIR))

from src.config import config
from src.models import ChartsMinuteData
from src.data import load_min_data
```

```python
dict_path = ROOT_DIR / "data" / "0014_50_pct_gappers_20031008_20250606.feather"
min_path = dict_path.with_name(dict_path.stem + "_min_data.feather")
chart = ChartsMinuteData(dict_path, min_path)
```

```python
df,metadata = chart.load_chart(12)
df["time"] = pd.to_datetime(df["time"])
df.set_index("time", inplace=True)
df_test = df.between_time("09:30", "16:00", inclusive="left")
# Group by date and get the first (smallest) and last (biggest) index for each date
date_groups = df_test.groupby(df_test.index.date)
biggest_indices = [group.index[-1] for _, group in date_groups]
smallest_indices = [group.index[0] for _, group in date_groups]

biggest_indices
```

```python
from src.models import ChartsWMOverride
from lightweight_charts import JupyterChart

# lw_chart = ChartsWMOverride()
lw_chart = JupyterChart()
lw_chart.set(df)


for start_time in smallest_indices:
    lw_chart.vertical_span(start_time, color="rgba(255, 255, 255, 0.5)")
for end_time in biggest_indices:
    lw_chart.vertical_span(end_time, color="rgba(255, 255, 0, 0.5)")

lw_chart.load()
```
